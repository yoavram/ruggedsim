## Fixation probability

```python
bs.warnings.simplefilter("ignore", bs.InstabilityWarning)
```

```python
estimate_rho = lambda data: np.mean(data)
estimate_ci = lambda arr: tuple(bs.ci(arr, estimate_rho, output='errorbar', n_samples=1000))

def fixation_cm_exact(T,U,u,s,H):
    return 2 * s * H / (1 + s * H)
fixation_sim_exact = fixation_cm_exact
def fixation_sim_e_exact(T,U,u,s,H):
    return fixation_cm_exact(T,U,u,s,H) * (1 + U * (T - 1)/(s * H))

def fixation_regression(df, PI, fixation_function, ax=None, color="k", annotate=True):
    xs = df[df.pi==PI].tau
    ys = df[df.pi==PI].fix
    
    # GLM: general linear model regression
    X = sm.add_constant(xs, prepend=True)
    glm = sm.GLM(ys, X, family=sm.families.Binomial())
    glm_results = glm.fit()
    b,a = glm_results.params
    pval = glm_results.pvalues[1]
    
    # estimare rho and use bootstrap to calculate CI
    taus = xs.unique()    
    taus.sort()
    rhos = df[df.pi==PI].groupby('tau').fix.agg(estimate_rho)
    errs = zeros(taus.shape)
    errs = df[df.pi==PI].groupby('tau').fix.agg(estimate_ci)
    errs = array([[x[0][0],x[1][0]] for x in errs.tolist()]).transpose()
    
    if ax == None:
        fig = figure(figsize=(6,4))
        ax = fig.add_subplot(111)
    
    ax.plot(taus, expit(b + a * taus), '-', color=color, linewidth=2)
    ax.errorbar(taus, rhos, yerr=errs, fmt='.k', elinewidth=2, capsize=5)
    ax.plot(taus, [fixation_function(TT,U,u,s,H) for TT in taus],'--', linewidth=2, color=color)
    if annotate:
        ax.annotate('slope=%.4f, intercept=%.2f, P=%.2g' % (a,b,pval),(0.02,0.02), 
                    xycoords='axes fraction',fontsize=2*annotate_fontize/3)
    plt.setp(ax.get_xticklabels(), fontsize=annotate_fontize)
    plt.setp(ax.get_yticklabels(), fontsize=annotate_fontize)
    return ax,a,b,pval
```

```python
annotate_xpos = 1
if s == 0.05:
    annotate_ypos = 0.23
if s == 0.1:
    annotate_ypos = 0.38

fig = figure(figsize=(10.5,9))

ax1 = fig.add_subplot(321)
fixation_regression(df, 10, fixation_cm_exact, ax1, red)
ax1.set_xticklabels([])
ax1.text(annotate_xpos,annotate_ypos,"CM",color=red, fontsize=line_label_fontsize)

ax2 = fig.add_subplot(323)
fixation_regression(df, 1.0, fixation_sim_exact, ax2, blue)
ax2.set_xticklabels([])
ax2.set_ylabel(r"Fixation probability - $\rho$")
ax2.text(annotate_xpos,annotate_ypos,"SIM",color=blue, fontsize=line_label_fontsize)

ax3 = fig.add_subplot(325)
fixation_regression(df, 1.01, fixation_sim_e_exact, ax3, green)
ax3.set_xticks([1]+range(10,51,10))
ax3.set_xticklabels([1]+range(10,51,10))
ax3.set_xlabel(r"Mutation rate fold increase - $\tau$")
ax3.text(annotate_xpos,annotate_ypos,r"SIM$_e$",color=green, fontsize=line_label_fontsize)

ax4 = fig.add_subplot(322)
fixation_regression(df_unloaded, 10, fixation_cm_exact, ax4, red)
ax4.set_xticklabels([])
ax4.set_yticklabels([])

ax5 = fig.add_subplot(324)
fixation_regression(df_unloaded, 1.0, fixation_sim_exact, ax5, blue)
ax5.set_xticklabels([])
ax5.set_yticklabels([])

ax6 = fig.add_subplot(326)
fixation_regression(df_unloaded, 1.01, fixation_sim_e_exact, ax6, green)
ax6.set_xticks([1]+range(10,51,10))
ax6.set_xticklabels([1]+range(10,51,10))
ax6.set_xlabel(r"Mutation rate fold increase - $\tau$")
ax6.set_yticklabels([])

plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), xlim=(0,51))
plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ymargin=0)
if s == 0.05:
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ylim=(0.1,0.25))
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), yticks=arange(0.15,0.25,0.05))
if s == 0.1:
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ylim=(0.25,0.4))
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), yticks=arange(0.3,0.39,0.05))

fig.subplots_adjust(left=None, bottom=0, right=None, top=None, wspace=None, hspace=0.25)
fig.tight_layout(pad=0)
```

## Adaptation rate

```python
estimate_Nq = lambda apps: 1 / apps.mean()

def estimate_nu(apps,fix):
    return estimate_Nq(apps) * estimate_rho(fix)
```

```python
estimate_nu_ci = lambda app_time, fix: array(bs.ci((app_time, fix), estimate_nu, n_samples=1000))
estimate_nu_ci_low = lambda app_time, fix: estimate_nu_ci(app_time, fix)[0]
estimate_nu_ci_high = lambda app_time, fix: estimate_nu_ci(app_time, fix)[1]
```

```python
grouped = df.groupby(('pi','tau'))
def f(group):
    return pd.Series({'nu' : estimate_nu(group.app_time, group.fix),
       'nu_ci_low' : estimate_nu_ci_low(group.app_time, group.fix),
       'nu_ci_high' : estimate_nu_ci_high(group.app_time, group.fix)})
applied = grouped.apply(f).reset_index()
```

```python
adaptation_rate_cm = lambda T,U,u,s,H: 4*N*H*(u**2)*(T**2)
adaptation_rate_sim = lambda T,U,u,s,H: 4*N*H*(u**2)*2*T/(2-T*U)
adaptation_rate_sim_e = lambda T,U,u,s,H: 4*N*H*(u**2)*(T**2)*(1 + U*(T-1)/(s*H))

series10 = applied[applied.pi==10]
series1 = applied[applied.pi==1]
series101 = applied[applied.pi==1.01]

fig = figure(figsize=(10,8), dpi=100)
ax = fig.add_subplot(111)

ax.errorbar(series10.tau,  series10.nu,  yerr=[series10.nu - series10.nu_ci_low, series10.nu_ci_high - series10.nu], 
              label="CM", color=red, fmt='o', capsize=5, elinewidth=2)
ax.plot(taus, [adaptation_rate_cm(TT,U,u,s,H) for TT in taus], '-', linewidth=2, color=red)
ax.errorbar(series1.tau,   series1.nu,   yerr=[series1.nu - series1.nu_ci_low, series1.nu_ci_high - series1.nu], 
              label="SIM", color=blue, fmt='s', capsize=5, elinewidth=2)
ax.plot(taus, [adaptation_rate_sim(TT,U,u,s,H) for TT in taus], '-', linewidth=2, color=blue)
ax.errorbar(series101.tau, series101.nu, yerr=[series101.nu - series101.nu_ci_low, series101.nu_ci_high - series101.nu], 
              label="SIMe", color=green, fmt='^', capsize=5, elinewidth=2)
ax.plot(taus, [adaptation_rate_sim_e(TT,U,u,s,H) for TT in taus], '--', linewidth=2, color=green)

ax.text(0.92, 2e-8, "NM", color='k', fontsize=line_label_fontsize)
ax.text(40, 2.8e-5, "CM", color=red, fontsize=line_label_fontsize)
ax.text(40, 6e-7, "SIM", color=blue, fontsize=line_label_fontsize)
ax.text(38, 2e-4, "SIMe", color=green, fontsize=line_label_fontsize)

ax.set_ylim((1.5e-8,5e-4))
ax.set_xlim((0.9,60))

ax.set_yscale("log")
ax.set_xscale("log")
ax.minorticks_off()

ax.set_xticks([1,3,10,50])
ax.set_xticklabels([1,3,10,50])

ax.set_ylabel(r"Adaptation rate - $\log{\nu}$")
ax.set_xlabel(r"Mutation rate fold increae - $\log{\tau}$")

fig.tight_layout()
```
