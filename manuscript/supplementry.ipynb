## Fixation probability - Fig. S2

```python
def expit_params(logit_b0, logit_b1):
    b0 = expit(logit_b0)
    b1 = expit(logit_b1 + logit_b0) - expit(logit_b0)
    return b0,b1
def logit_params(b0, b1):
    logit_b0 = logit(b0)
    logit_b1 = logit(b0 + b1) - logit(b0)
    return logit_b0, logit_b1
```

```python
def fixation_cm_exact(T,U,u,s,H):
    return 2 * s * H / (1 + s * H)
fixation_sim_exact = fixation_cm_exact
def fixation_sim_e_exact(T,U,u,s,H):
    return fixation_cm_exact(T,U,u,s,H) * (1 + U * (T - 1)/(s * H))
```

```python
PI = 10
xs = df_unloaded[df_unloaded.pi==PI].tau
ys = df_unloaded[df_unloaded.pi==PI].fix

# GLM: general linear model regression
X = sm.add_constant(xs, prepend=True)
glm = sm.GLM(ys, X, family=sm.families.Binomial())
glm_results = glm.fit()
b,a = glm_results.params
ci = glm_results.conf_int(alpha=0.001)
low,high = ci[0][1],ci[1][1]
print low,high
H0_no_slope = low < 0.0 < high
sim_e_slope = logit_params(fixation_sim_e_exact(0,U,u,s,H), fixation_sim_e_exact(1,U,u,s,H)-fixation_sim_e_exact(0,U,u,s,H))[1]
H0_sim_e_slope = low < sim_e_slope < high
print "H0_no_slope %r, H0_sim_e_slope %r" % (H0_no_slope, H0_sim_e_slope)
print "slope:", expit_params(b,a), (2*s*H/(1+s*H), 2*U/(1+s*H)), (fixation_sim_e_exact(0,U,u,s,H), fixation_sim_e_exact(1,U,u,s,H)-fixation_sim_e_exact(0,U,u,s,H))
```

```python
estimate_rho = lambda data: np.mean(data)
estimate_ci = lambda arr: tuple(bs.ci(arr, estimate_rho, output='errorbar', n_samples=10000))

def fixation_regression(df, PI, fixation_function, ax=None, color="k", annotate=False, alpha=0.05/6):
    xs = df[df.pi==PI].tau
    ys = df[df.pi==PI].fix
    
    # GLM: general linear model regression
    X = sm.add_constant(xs, prepend=True)
    glm = sm.GLM(ys, X, family=sm.families.Binomial())
    glm_results = glm.fit()
    b,a = glm_results.params
    ci = glm_results.conf_int(alpha=alpha)
    low,high = ci[0][1],ci[1][1]
    H0_no_slope = low < 0.0 < high
    sim_e_slope = logit_params(fixation_sim_e_exact(0,U,u,s,H), fixation_sim_e_exact(1,U,u,s,H)-fixation_sim_e_exact(0,U,u,s,H))[1]
    H0_sim_e_slope = low < sim_e_slope < high
    print "H0_no_slope %r, H0_sim_e_slope %r" % (H0_no_slope, H0_sim_e_slope)
    
    # estimare rho and use bootstrap to calculate CI
    taus = xs.unique()    
    taus.sort()
    rhos = df[df.pi==PI].groupby('tau').fix.agg(estimate_rho)
    errs = zeros(taus.shape)
    errs = df[df.pi==PI].groupby('tau').fix.agg(estimate_ci)
    errs = array([[x[0][0],x[1][0]] for x in errs.tolist()]).transpose()
    
    if ax == None:
        fig = figure(figsize=(6,4))
        ax = fig.add_subplot(111)
    
    ax.plot(taus, expit(b + a * taus), '-', color=color, linewidth=2)
    ax.errorbar(taus, rhos, yerr=errs, fmt='.k', elinewidth=2, capsize=5)
    ax.plot(taus, [fixation_function(TT,U,u,s,H) for TT in taus],'--', linewidth=2, color=color)
    if annotate:
        ax.annotate('slope=%.4f, intercept=%.2f, \n H0 no slope %r, H0 SIMe slope %r' % (a,b,H0_no_slope,H0_sim_e_slope),(0.02,0.02), 
                    xycoords='axes fraction',fontsize=2*annotate_fontize/3)
    plt.setp(ax.get_xticklabels(), fontsize=annotate_fontize)
    plt.setp(ax.get_yticklabels(), fontsize=annotate_fontize)
    return 
```

```python
annotate_xpos = 1
if s == 0.05:
    annotate_ypos = 0.23
if s == 0.1:
    annotate_ypos = 0.38

figs2 = figure(figsize=(10.5,9))

ax1 = figs2.add_subplot(321)
fixation_regression(df, 10, fixation_cm_exact, ax1, colors["CM"])
ax1.set_xticklabels([])
ax1.text(annotate_xpos,annotate_ypos,"CM",color=red, fontsize=line_label_fontsize)

ax2 = figs2.add_subplot(323)
fixation_regression(df, 1.0, fixation_sim_exact, ax2, colors["SIM"])
ax2.set_xticklabels([])
ax2.set_ylabel(r"Fixation probability - $\rho$")
ax2.text(annotate_xpos,annotate_ypos,"SIM",color=blue, fontsize=line_label_fontsize)

ax3 = figs2.add_subplot(325)
fixation_regression(df, 1.01, fixation_sim_e_exact, ax3, colors["NM"])
ax3.set_xticks([1]+range(10,51,10))
ax3.set_xticklabels([1]+range(10,51,10))
#ax3.set_xlabel(r"Mutation rate fold increase - $\tau$")
ax3.text(annotate_xpos,annotate_ypos,r"SIM$_e$",color=green, fontsize=line_label_fontsize)

ax4 = figs2.add_subplot(322)
fixation_regression(df_unloaded, 10, fixation_cm_exact, ax4, colors["CM"])
ax4.set_xticklabels([])
ax4.set_yticklabels([])

ax5 = figs2.add_subplot(324)
fixation_regression(df_unloaded, 1.0, fixation_sim_exact, ax5, colors["SIM"])
ax5.set_xticklabels([])
ax5.set_yticklabels([])

ax6 = figs2.add_subplot(326)
fixation_regression(df_unloaded, 1.01, fixation_sim_e_exact, ax6, colors["NM"])
ax6.set_xticks([1]+range(10,51,10))
ax6.set_xticklabels([1]+range(10,51,10))
#ax6.set_xlabel(r"Mutation rate fold increase - $\tau$")
ax6.set_yticklabels([])

plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), xlim=(0,51))
plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ymargin=0)
if s == 0.05:
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ylim=(0.1,0.25))
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), yticks=arange(0.15,0.25,0.05))
if s == 0.1:
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), ylim=(0.25,0.4))
    plt.setp((ax1,ax2,ax3,ax4,ax5,ax6), yticks=arange(0.3,0.39,0.05))

figs2.subplots_adjust( hspace=0.25)
figs2.text(0.3,0.04,r"Mutation rate fold increase - $\tau$",fontsize=style.lib['ggplot']['axes.labelsize'],color=style.lib['ggplot']['axes.labelcolor'])
```

```python
estimate_Nq = lambda apps: 1 / apps.mean()

def estimate_nu(apps,fix):
    return estimate_Nq(apps) * estimate_rho(fix)
```

```python
estimate_nu_ci = lambda app_time, fix: array(bs.ci((app_time, fix), estimate_nu, n_samples=1000))
estimate_nu_ci_low = lambda app_time, fix: estimate_nu_ci(app_time, fix)[0]
estimate_nu_ci_high = lambda app_time, fix: estimate_nu_ci(app_time, fix)[1]
```

## Mean fitness at MSB - Fig. 3

```python
from scipy.stats import poisson
from scipy.linalg import eigvals

def mean_fitness(T,U,u,s,G=25,PI=1):
    w = (1 - s)**arange(G)
    M = repeat(w,G).reshape(G,G).transpose()
    for i in arange(G):
        for j in arange(G):
            if j >= PI:
                tau = T
            else:
                tau = 1
            #print j,'->',i,':',
            if j <= i:
                M[i,j] *= poisson(tau * U).pmf(i - j)
            if j >= i:
                M[i,j] *= poisson(tau * u).pmf(j - i)
    return max(eigvals(M)).real
```

[Colors](http://colorbrewer2.org/index.php?type=sequential&scheme=YlGn&n=9) from [ColorBrewer](http://colorbrewer2.org/)

```python
colors2 = '0xFFFFE5; 0xFFF7BC; 0xFEE391; 0xFEC44F; 0xFE9929; 0xEC7014; 0xCC4C02; 0x993404; 0x662506; 0x331608;'
colors2 = colors2.split(';')
colors2 = [c.replace('0x','#').strip() for c in colors2[:-1]]
print colors2
```

```python
class Genotype:
    def __init__(self, Aa, Bb, num_deleterious=0):
        self.Aa = Aa
        self.Bb = Bb
        self.num_deleterious = num_deleterious
        self.name = Aa+Bb
        if num_deleterious > -1:
            self.name += '/' + str(num_deleterious)
        if Aa == 'A' and Bb == 'B':
            self.fitness = (1 - s) ** num_deleterious * (1 + s * H)
        elif Aa == 'A' or Bb == 'B':
            self.fitness = (1 - s) ** (num_deleterious + 1)
        else:
            self.fitness = (1 - s) ** num_deleterious
        self.color = '#ffffff'
        self.sons = []
        if num_deleterious < G:
            self.sons.append(Genotype(Aa, Bb, num_deleterious + 1))
        if Aa.islower():
            self.sons.append(Genotype(Aa.upper(), Bb, num_deleterious))
        if Bb.islower():
            self.sons.append(Genotype(Aa, Bb.upper(), num_deleterious))
    def __repr__(self):
        return self.name
    
    def set_color(self, color):
        self.color = color
        
    def to_node(self):
        style = 'filled'
        shape = "ellipse"
        if self.fitness >= 1: 
            style = '"' + style + '"'
            shape = "rect"
        if self.fitness == 1 + s * H:
            style = '"diagonals,' + style[1:]
        node = pydot.Node(self.name, style=style, fillcolor=self.color, fontname=fontname, shape=shape)
        if self.fitness < (1 - s) ** 3: 
            node.set_fontcolor('#CCCC9A')
        return node
    
    def __eq__(self, other):
        return self.name == other.name
    
    def all_paths(self):
        if len(self.sons) == 0:
            return [self]
        sibs = self.sons + [self]
        for g in self.sons:            
            sibs.extend([s for s in g.all_paths() if s not in sibs])
        return sibs
    
    def is_leaf(self):
        return self.num_deleterious == G
```

### Stochastic model - panel b

```python
ab = Genotype('a','b',0)
genotypes = sorted(ab.all_paths(), key=lambda x: x.name)
fitness = sorted(unique([g.fitness for g in genotypes]), reverse=True)
print len(fitness)
print fitness
fitness_to_color = { w:colors2[i] for i,w in enumerate(fitness) }    
for g in genotypes:
    g.set_color(fitness_to_color[g.fitness])
```

```python
graph = pydot.Dot(graph_type='digraph',strict=True)
for g in genotypes:
    node = g.to_node()
    graph.add_node(node)
    
for g in genotypes:
    if g.is_leaf():
        node = pydot.Node(g.name + '...')
        node.set_label('...')
        node.set_style('filled')
        node.set_fillcolor(colors2[-1])
        node.set_fontcolor('#CCCC9A')
        node.set_fontname(fontname)
        graph.add_node(node)
        edge = pydot.Edge(g.to_node(), node)
        edge.set_label(' U')
        edge.set_style('dashed')
        edge.set_fontname(fontname)
        graph.add_edge(edge)
    for gs in g.sons:
        edge = pydot.Edge(g.to_node(), gs.to_node())
        if g.name[:2] == gs.name[:2]:
            edge.set_label(' U')
            edge.set_style('dashed')

        else:
            edge.set_label(u" \u03BC")
        edge.set_fontname(fontname)
        graph.add_edge(edge)
graph.set_ratio(2)
graph
```

```python
def no_del(node):
    node.set_name(node.get_name()[1:3])
    return node
```

```python
graph = pydot.Dot(graph_type='digraph',strict=True)

rip = pydot.Node("RIP", style="filled", fillcolor=colors2[-3], fontname=fontname)
graph.add_node(rip)

for g in [x for x in genotypes if x.name=='ab/0' or  x.name=='aB/0' or  x.name=='Ab/0' or  x.name=='AB/0']:
    graph.add_node(no_del(g.to_node()))

def add_benef(src, dst, label=u" \u03BC", hp="_"):
    edge = pydot.Edge(src, dst)
    edge.set_label(label)
    edge.set_fontname(fontname)
    edge.set_headport(hp)
    graph.add_edge(edge)
    return edge
    
def add_del(src):
    edge = pydot.Edge(src, rip)
    edge.set_label(' U')
    edge.set_style('dashed')
    edge.set_fontname(fontname)
    graph.add_edge(edge)
    return edge
    
add_benef('ab','aB')
add_benef('ab','Ab')
add_benef('ab','AB', label=u"<&nbsp;\u03BC<SUP>2</SUP>>", hp='n')
add_benef('aB','AB')
add_benef('Ab','AB')

add_del('ab')
add_del('aB')
add_del('Ab')
add_del('AB')

graph.set_ratio(1)
graph
```

```python
BOTH_SIDES = False

def make_bar(ax, ys, yerrs, color, title, alpha=1):
    f = ax.bar(left=range(len(ys.tau2)), height=ys.invader-0.5, bottom=0.5, yerr=yerrs.invader, 
           align="center", width=0.5, color=color, alpha=alpha,
           error_kw={"elinewidth":2, "capsize":5, "zorder":3,"ecolor":'k'})
    ax.set_xticks(range(len(ys.tau2)))
    ax.set_xticklabels([int(x) for x in ys.tau2])
    ax.set_yticklabels([])
    ax.set_xlim((-0.75,len(ys.tau2)-0.25))
    ax.minorticks_off()
    ax.set_ylim((-0.1,1.1))
    ax.set_title(title,fontsize=line_label_fontsize, color=rcParams['axes.labelcolor'])

invader = df_compete.groupby(('pi1','tau1','pi2','tau2')).invader
ys = invader.agg(np.mean).reset_index()
errs = (invader.agg(lambda arr: np.std(arr)) / np.sqrt(invader.agg(np.size))).reset_index()

w,h = rcParams['figure.figsize']
fig5 = figure(figsize=(w*1.5,h*0.75))
ax = [None]*4
ax[0] = plt.subplot2grid((1,7), (0,0))
ax[1] = plt.subplot2grid((1,7), (0,1), colspan=2)
ax[2] = plt.subplot2grid((1,7), (0,3), colspan=2)
ax[3] = plt.subplot2grid((1,7), (0,5), colspan=2)

fig5.subplots_adjust(wspace=0)

crit = (ys.pi1==0) & (ys.pi2==0)
make_bar(ax[0], ys[crit], errs[crit], colors["NM"], "NM vs. NM")
ax[0].set_xticklabels([])

if BOTH_SIDES:
    crit = (ys.pi2==0) & (ys.pi1==1)
    make_bar(ax[1], ys[crit], errs[crit], colors["NM"], "SIM vs. NM")
crit = (ys.pi1==0) & (ys.pi2==1)
make_bar(ax[1], ys[crit], errs[crit], colors["SIM"], "SIM vs. NM")

if BOTH_SIDES:
    crit = (ys.pi2==0) & (ys.pi1==10)
    make_bar(ax[2], ys[crit], errs[crit], colors["NM"], "CM vs. NM")
crit = (ys.pi1==0) & (ys.pi2==10)
make_bar(ax[2], ys[crit], errs[crit], colors["CM"], "CM vs. NM")

if BOTH_SIDES:
    crit = (ys.pi2==10) & (ys.pi1==1)
    make_bar(ax[3], ys[crit], errs[crit], colors["CM"], "SIM vs. CM")
crit = (ys.pi1==10) & (ys.pi2==1)
make_bar(ax[3], ys[crit], errs[crit], colors["SIM"], "SIM vs. CM")


ax[0].set_ylabel("Final frequency", fontsize=line_label_fontsize)
ax[0].set_yticks(arange(0,1.1,0.25))
ax[0].set_yticklabels(arange(0,1.1,0.25))

txt1 = plt.figtext(0.6,0,r"Mutation rate fold increae - $\tau$",horizontalalignment='center', verticalalignment='center',
    fontsize=line_label_fontsize, color=rcParams['axes.labelcolor'])

fig5.tight_layout()
```

```python
df_fix_seq10U_success = df_fix_seq10U[df_fix_seq10U.fix==True]
len(df_fix_seq10U), len(df_fix_seq10U_success)
```

```python
grouped = df_fix_seq10U_success.groupby(('pi','tau','s','U','H'))
def f(group):
    return pd.Series({
        'W_obs' : mean(group.fix_W),
        'W_app':  mean(group.app_W) * (1 + group.s.unique()[0] * group.H.unique()[0]),
        'W_msb':  mean(group.msb_W) * (1 + group.s.unique()[0] * group.H.unique()[0]),
        'W_exp_CM': exp(-group.tau.unique()[0] * group.U.unique()[0]) * (1 + group.s.unique()[0] * group.H.unique()[0]),
        'W_exp_NM': exp(-group.U.unique()[0]) * (1 + group.s.unique()[0] * group.H.unique()[0]),
        'W_no_hh': mean(group[group.app_best==0].fix_W),
        'W_hh': mean(group[group.app_best!=0].fix_W),
        'W_no_mr': mean(group[group.app_best==group.fix_best].fix_W),
        'W_mr': mean(group[group.app_best!=group.fix_best].fix_W),
        'ratchet_clicked': (group.app_best < group.fix_best).mean(),
        'mean_best': group.app_best.mean(),
        'samples': len(group.app_best)       
                      })
applied = grouped.apply(f).reset_index()
```

```python
figX = figure()
ax = figX.add_subplot(111)

ax.errorbar(applied[applied.pi==1].tau, applied[applied.pi==1].W_obs, fmt='o', color=colors["SIM"])
ax.errorbar(applied[applied.pi==10].tau, applied[applied.pi==10].W_obs, fmt='o', color=colors["CM"])

# perfect fit
#ax.errorbar(applied[applied.pi==1].tau, applied[applied.pi==1].W_app, fmt='v', color=colors["SIM"])
#ax.errorbar(applied[applied.pi==10].tau, applied[applied.pi==10].W_app, fmt='v', color=colors["CM"])

# perfect fit
#ax.errorbar(applied[applied.pi==1].tau, applied[applied.pi==1].W_msb, fmt='v', color=colors["SIM"])
#ax.errorbar(applied[applied.pi==10].tau, applied[applied.pi==10].W_msb, fmt='v', color=colors["CM"])

ax.errorbar(applied[applied.pi==1].tau, applied[applied.pi==1].W_no_hh, fmt='v', color=colors["SIM"])
ax.errorbar(applied[applied.pi==10].tau, applied[applied.pi==10].W_no_hh, fmt='v', color=colors["CM"])

ax.errorbar(applied[applied.pi==1].tau, applied[applied.pi==1].W_no_mr, fmt='^', color=colors["SIM"], markerfacecolor='white')
ax.errorbar(applied[applied.pi==10].tau, applied[applied.pi==10].W_no_mr, fmt='^', color=colors["CM"], markerfacecolor='white')

ax.plot(applied[applied.pi==1].tau, applied[applied.pi==1].W_exp_NM, '-', lw=2, color=colors["SIM"], label="SIM")
ax.plot(applied[applied.pi==10].tau, applied[applied.pi==10].W_exp_CM, '-', lw=2, color=colors["CM"], label="CM")

ax.set_xlim(-1,51)
#ax.set_ylim(0.8, 1.11)

ax.legend(loc=3)
ax.set_title("s=%.3f U=%.3f" % (s,U))
ax.set_xlabel(r"Mutation rate fold increae - $\log{\tau}$", fontsize=line_label_fontsize)
ax.set_ylabel("Avg. mean fitness post adaptation", fontsize=line_label_fontsize-4)

ax.text(12, 0.81, "o - Observed\nv - No deleterious background\n^ - No Muller's Ratchet\n-- - MSB")

figX.tight_layout()
```
