# MSB analysis

```python
font = {'family' : 'sans-serif',
        'size'   : 12}
matplotlib.rc('font', **font)
rcParams['figure.figsize'] = 8,6
```

```python
import cPickle
import gzip
import pandas as pd
```

```python
jobname = 'msb'
U = 0.0004
```

```python
with gzip.open( "all_W_%s.pickle.gz" % jobname, "r") as f:
    all_W = cPickle.load(f)
```

```python
with gzip.open( "all_data_%s.pickle.gz" % jobname, "r") as f:
    all_data = cPickle.load(f)
```

```python
def find_in_data(fname):
    for d in all_data:
        if d['fname'] == fname:
            return d
```

```python
W200 = [W[200] for W in all_W.values()]
```

```python
scatter(range(len(W200)), log(W200))
axhline(y=-U)
axhline(y=0)
ylim((-2*U,0))
```

```python
W_sim = []
for fname, W in all_W.items():
    d = find_in_data(fname)
    if d['pi'] != 1:
        continue
    W_sim.extend(W[200:])
```

```python
scatter(range(len(W_sim)), log(W_sim))
axhline(y=-U)
ylim((-1.1*U,-0.9*U))
```

```python
df = []
for fname,W in all_W.items():
    d = find_in_data(fname)
    d['mean_W'] = mean(W[200:])
    d['se_W'] = std(W[200:]) / sqrt(len(W)-200)
    df.append(d)
```

```python
W_sim = [d['mean_W'] for d in df if d['pi'] == 1]
```

```python
scatter(range(len(W_sim)), log(W_sim))
axhline(y=-U)
ylim((-1.02*U,-0.98*U))
```

```python
df = pd.DataFrame(df)
```

```python
df.to_csv("msb.csv")
```
