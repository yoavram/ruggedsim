### Appearance probability

Disregarding back mutations, the general form of the probability of appearance of a double mutant is:
$$
q = f_{ab} \omega_{ab} \mu_{ab}^2 + f_{Ab} \omega_{Ab} \mu_{Ab} + f_{aB} \omega_{aB} \mu_{aB}
$$
Assuming the effect of deleterious mutations is $s$ and that the population is at a mutation-selection balance then we can subtitute the frequencies $f_{ab} = e^{-\frac{U_{ab}}{s}}, f_{Ab}=f_{aB}=\frac{\mu_{ab}}{s}e^{-\frac{\mu_{ab}}{s}}$.
Fitness takes to account for both the Fisherian fitness of the individuals ($\omega_{ab}=1, \omega_{Ab}=\omega_{aB}=1-s$) and the fitness cost of mutation (the mutational load), given by the probability that no new mutation occurs $e^{-U_{k}}$:
$$
q = e^{-\frac{U_{ab}}{s}} e^{-U_{ab}} \mu_{ab}^2 + 2 \frac{\mu_{ab}}{s}e^{-\frac{\mu_{ab}}{s}} (1-s) e^{-U_{Ab}} \mu_{Ab}
$$

#### CM
In this case $U_{k}=\tau U, \mu_{k}=\tau \mu$:
$$
q_{CM} = e^{-\frac{\tau U}{s}} e^{-\tau U} (\tau \mu)^2 + 2\frac{\tau \mu}{s} e^{-\frac{\tau \mu}{s}} e^{-\tau U} \tau \mu \Rightarrow \\\\
q_{CM} = e^{-\tau U} (\tau \mu)^2 ( e^{-\frac{\tau U}{s}} + \frac{2}{s}  e^{-\frac{\tau \mu}{s}})
$$

```python
def appearance_cm_exact(T,U,u,s,H):
    return e**(-T*U) * (T*u)**2 * ( e**(-T*U/s) + 2 * e**(-T*u/s) / s)
```

#### SIM
Here, $U_{ab}=U, U_{aB}=U{Ab}=\tau U, \mu_{ab}=\mu, \mu_{aB}=\mu_{Ab}=\tau \mu$:
$$
q_{SIM} = e^{-\frac{U}{s}} e^{-U} \mu^2 + 2\frac{\mu}{s} e^{-\frac{\mu}{s}} e^{-\tau U} \tau \mu \Rightarrow \\\\
q_{SIM} = \mu^2 (e^{-U(1+\frac{1}{s})} + \frac{2\tau}{s} e^{-\frac{\mu}{s}-\tau U})
$$

```python
def appearance_sim_exact(T,U,u,s,H):
    return u**2 * (e**(-U*(1+1/s)) + (2*T/s) * e**((-u/s)*T*U)) 
```

```python
cm_sim_nm(appearance_cm_exact,appearance_sim_exact,"appearance probability")
```

### Fixation probability
#### CM

$$
\alpha = \frac{(1+sH)e^{-\tau U}}{\bar{\omega}} \\\\
\bar{\omega} = e^{-\tau U} \Rightarrow \alpha = 1+sH\\\\
\pi = 2\frac{\alpha - 1}{\alpha} \Rightarrow \\\\
\pi = 2 \frac{sH}{1+sH}
$$

```python
def fixation_cm_exact(T,U,u,s,H):
    return 2 * s*H / (1 + s*H)
```

#### SIM

$$
\alpha_{SIM} = \frac{(1+s H)e^{-U}}{\bar{\omega}_{SIM}} \\\\
\bar{\omega}_{SIM} = e^{-U/s} + (1-e^{-U/s})\sum_{k \ge 0} {(1-s)^{k+1} e^{-\tau U/s}(\tau U/s)^k/k!} = \\\\
e^{-U/s} + (1-e^{-U/s}) (1-s) e^{-\tau U/s} \sum_{k \ge 0} {( (1-s) \tau U/s)^k / k!)} = \\\\
e^{-U/s} + (1-e^{-U/s}) (1-s) e^{-\tau U/s} e^{(1-s)\tau U /s} = \\\\
e^{-U/s} + (1-e^{-U/s}) (1-s) e^{-\tau U/s + \tau U/s -\tau U} = \\\\
e^{-U/s} + (1-e^{-U/s}) (1-s) e^{-\tau U} \Rightarrow \\\\
\bar{\omega}_{SIM} = e^{-U/s} + (1-e^{-U/s})(1-s)e^{-\tau U} \Rightarrow \\\\
\alpha_{SIM} = \frac{1+sH}{e^{U-U/s}+(1-e^{-U/s})(1-s)e^{U-\tau U}} \Rightarrow \\\\
\pi_{SIM} = 2\frac{1+sH-e^{U-U/s}-(1-e^{-U/s})(1-s)e^{U-\tau U}}{{1+sH}} \Rightarrow \\\\
\pi_{SIM} = 2-2\frac{e^{U-U/s} + (1-e^{-U/s})(1-s)e^{U-\tau U}}{{1+sH}}
$$

```python
def fixation_sim_exact(T,U,u,s,H):
    return 2 - 2 * (e**(U-U/s) + (1 - e**(-U/s)) * (1-s) * e**(U-T*U) ) / (1+s*H)
```

```python
cm_sim_nm(fixation_cm_exact,fixation_sim_exact,"fixation probability")
```

### Adaptation rate

In general, the waiting time for appearance can be approximated by $1/Nq$ and therefore the waiting time for the appearance of a double mutant that will fixate is $1/Nq\pi$. The adaptation time is close to this waiting time, as fixation is a fast process in comparison with waiting. The adaptation rate is the inverse of the adaptation time.

```python
def adaptation_rate_cm_exact(T,U,u,s,H):
    return N * appearance_cm_exact(T,U,u,s,H) * fixation_cm_exact(T,U,u,s,H)
def adaptation_rate_sim_exact(T,U,u,s,H):
    return N * appearance_sim_exact(T,U,u,s,H) * fixation_sim_exact(T,U,u,s,H)
```

```python
cm_sim_nm(adaptation_rate_cm_exact, adaptation_rate_sim_exact, "Adaptation rate")
```

```python
print '%.8f' % adaptation_rate_sim_exact(T,U,u,s,H)
```

## Analytical approximation

Here we develop some analytical approximations to the above expressions.
This approximations make use of Tayler expansions and removing of small terms from sums.
We also compare this approximations to the exact expressions defined above.

```python
def compare_exact_approx(exact,approx,color='k'):
    fig = figure(figsize=(14,6))
    ax=subplot(1,3,1)
    taus = logspace(0,2.5)
    ax.plot(taus,[abs(approx(TT,U,u,s,H)-exact(TT,U,u,s,H))/exact(TT,U,u,s,H) for TT in taus], color, linewidth=2.5);
    ax.set_xlabel(r"$\tau$")
    ax.set_ylabel("Relative error")
    ax.set_xscale("log")
    ax.set_xlim(min(taus),max(taus))
    ax.axhline(y=1)
    #ax.set_ylim(0,1)
    
    ax=subplot(1,3,2)
    ses = logspace(-3,-1)
    ax.plot(ses,[abs(approx(T,U,u,ss,H)-exact(T,U,u,ss,H))/exact(T,U,u,ss,H) for ss in ses], color, linewidth=2.5);
    ax.set_xlabel("s")
    #ax.set_ylabel("Relative error"); 
    ax.set_xscale("log")
    ax.axhline(y=1)
    #ax.set_ylim(0,1)
    
    ax=subplot(1,3,3)
    Us = logspace(-5,-1)
    ax.plot(Us,[abs(approx(T,UU,UU/5000,s,H)-exact(T,UU,UU/5000,s,H))/exact(T,UU,UU/5000,s,H) for UU in Us], color, linewidth=2.5);
    ax.set_xlabel("U")
    #ax.set_ylabel("Relative error")
    ax.set_xscale("log")
    ax.axhline(y=1)
    #ax.set_ylim(0,1);
    return fig
```

### Appearance probability
#### CM

$$
q_{CM} = e^{-\tau U} (\tau \mu)^2 ( e^{-\frac{\tau U}{s}} + \frac{2}{s}  e^{-\frac{\tau \mu}{s}}) \approx \\\\
(1-\tau U) (\tau \mu)^2 ( 1-\frac{\tau U}{s} +  \frac{2}{s} (1-\frac{\tau \mu}{s}) = \\\\
(1-\tau U) (\frac{\tau \mu}{s})^2 ( s^2-s\tau U +  2s - 2\tau \mu) \approx \\\\
(1-\tau U) (\frac{\tau \mu}{s})^2 ( 2s -s\tau U) = \\\\
(1-\tau U) ( 2 -\tau U) \frac{(\tau \mu)^2}{s}
$$

```python
def appearance_cm_approx(T,U,u,s,H):
    return (1 - T * U) * (2 - T * U) * ((T*u)**2) / s
```

```python
compare_exact_approx(appearance_cm_exact,appearance_cm_approx, color='r');
```

#### SIM
The approximation assumes that $\mu < U < \tau U << 1, s<<1$. 

$$
q_{SIM} = \mu^2 (e^{-U(1+\frac{1}{s})} + \frac{2\tau}{s} e^{-\frac{\mu}{s}-\tau U}) \approx \\\\
\mu^2 (1-U(1+\frac{1}{s}) + \frac{2\tau}{s} (1-\frac{\mu}{s}-\tau U)) \approx \\\\
\mu^2 (1-\frac{U}{s} + \frac{2\tau}{s} (1-\tau U)) = \\\\
\frac{\mu^2}{s} (s-U + 2\tau - 2\tau^2 U) \approx \\\\
\frac{\mu^2}{s} (2\tau - 2\tau^2 U) = \\\\
(1 - \tau U) \frac{2\tau \mu^2}{s}
$$

```python
def appearance_sim_approx(T,U,u,s,H):
    return (1 - T * U) * (2 * T * u**2) / s
```

```python
fig = compare_exact_approx(appearance_sim_exact,appearance_sim_approx,color='b');
```

### Fixation probability

#### CM/NM
$$
\pi_{CM} = \pi = 2 \frac{sH}{1+sH} \approx 2sH
$$

```python
def fixation_cm_approx(T,U,u,s,H):
    return 2*s*H
```

```python
compare_exact_approx(fixation_cm_exact,fixation_cm_approx,color='r');
```

#### SIM
Assuming $sH<<1, U < U/s < 1, \tau U < 1$,

$$
\pi_{SIM} = 2\frac{1+sH-e^{U-U/s}-(1-e^{-U/s})(1-s)e^{U-\tau U}}{1+sH} \approx \\\\
2(1+sH-e^{U-U/s}-(1-e^{-U/s})(1-s)e^{U-\tau U}) \approx \\\\
2(1+sH-(1+U-U/s)-(1-(1-U/s))(1-s)(1+U-\tau U)) = \\\\
2(sH-U+U/s-U/s(1-s)(1+U-\tau U)) = \\\\
2(sH-U+U/s-U/s(1-s)(1-U(\tau-1))) = \\\\
2(sH +\frac{U^2}{s}(\tau-1) - U^2(\tau-1) ) = \\\\
2sH + 2\frac{U^2}{s}(\tau - 1)(1 -s)
$$

```python
def fixation_sim_approx(T,U,u,s,H):
    return 2*s*H + 2*(U**2 /s)*(T-1)*(1-s)
```

```python
fig = compare_exact_approx(fixation_sim_exact,fixation_sim_approx,color='b');
```

```python
plot_filename = 'exact_approx_fixation_SIM_U_%.4f_s_%.2f_logN_%d' % (U,s,log10(N))
print(plot_filename)
fig.savefig('../manuscript/' + plot_filename + '.png')
```

### Adaptation rate
#### CM

$$
E[T_{CM}]^{-1} \approx N q_{CM} \pi_{CM}= \\\\
2 N s H (1-\tau U) ( 2 -\tau U) \frac{(\tau \mu)^2}{s} = \\\\
2 N H (\tau \mu)^2 (1-\tau U) ( 2 -\tau U)
$$

```python
def adaptation_rate_cm_approx(T,U,u,s,H):
    return 2*N*H * (T * u) ** 2 * (1-T*U)*(2-T*U)
```

```python
compare_exact_approx(adaptation_rate_cm_exact, adaptation_rate_cm_approx,color='r');
```

#### SIM

$$
E[T_{SIM}]^{-1} \approx N q_{SIM} \pi_{SIM} = \\\\
N (2sH + 2\frac{U^2}{s}(\tau - 1)(1 -s)) (1 - \tau U) \frac{2\tau \mu^2}{s} = \\\\
4 N \frac{\tau \mu^2}{s^2} (1 - \tau U) (s^2H + U^2(\tau - 1)(1 -s)) 
$$

```python
def adaptation_rate_sim_approx(T,U,u,s,H):
    return 4 * N * T * (u/s)  ** 2 * (1-T*U) * (s**2 * H + U**2 * (T-1) * (1-s))
```

```python
fig=compare_exact_approx(adaptation_rate_sim_exact, adaptation_rate_sim_approx,color='b');
```

```python
plot_filename = 'exact_approx_adaptation_SIM_U_%.4f_s_%.2f_logN_%d' % (U,s,log10(N))
print(plot_filename)
fig.savefig('../manuscript/' + plot_filename + '.png')
```

## Comparing CM and SIM

### Appearance probability

$$
\frac{q_{CM}}{q_{SIM}} \approx \\\\
\frac{(1-\tau U) ( 2 -\tau U) \frac{(\tau \mu)^2}{s}}{(1 - \tau U) \frac{2\tau \mu^2}{s}} = \\\\
(1 -\tau U) \tau/2
$$

Now, to check when this is larger/smaller than 1:
$$
(1 -\tau U) \tau/2 - 1 = 0 \Rightarrow \\\\
U \tau^2 - \tau + 2 = 0 \Rightarrow
$$ 

Since we are interested in $\tau>1$,
$$
\tau = \frac{1 + \sqrt{1-8U}}{2U} \approx \frac{1+1}{2U} = U^{-1}
$$

### Double mutant relative growth rate
#### Comparison of CM and SIM

$$
\alpha_{CM} = 1 + sH \\\\
\alpha_{SIM} = \frac{1+sH}{\bar{\omega}_{SIM}} \Rightarrow \\\\
\frac{\alpha_{CM}}{\alpha_{SIM}} = \bar{\omega}_{SIM} 
$$
Clearly, $\bar{\omega}_{SIM}<1$ as the fraction of the population with fitness $>1$ is negligible, and therefore 
$$
\alpha_{SIM} > \alpha_{CM} \Rightarrow 
\pi_{SIM} > \pi_{CM}
$$

The size of the advantage to SIM is $1/\bar{\omega}_{SIM}$:

$$
\bar{\omega}_{SIM} = e^{U-U/s}+(1-e^{-U/s})(1-s)e^{U-\tau U} = \\\\
e^U e^{-U/s} + e^U e^{-\tau U} (1-s) (1-e^{-U/s}) = \\\\
e^{-U(1-s)} (1 + e^{-\tau U} (1-s) (e^{U/s}-1))
$$

Therefore, increasing $\tau$ increases the advantage of SIM, but the dependence on *U* and *s* is more complicated

```python
def compare_simulation_to_analysis(simulation_mean, simulation_sed, exact, approx, mytitle=''):
    fig = figure()
    ax = fig.add_subplot(1,1,1)
    ax.errorbar(taus, simulation_mean, fmt='k.-.', label='Simulation', yerr=simulation_sed)
    ax.plot(taus, [1./exact(TT,U,u,s,H) for TT in taus],'b-',linewidth=2, label='Full approx.')
    ax.plot(taus, [1./approx(TT,U,u,s,H) for TT in taus],'r-.', linewidth=3, label='1st order approx.')
    ax.set_xlabel(r"Mutation rate increase - $\tau$")
    ax.set_ylabel('Waiting time')
    #ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim((min(taus),max(taus)))
    ax.set_title(mytitle)
    ax.legend(loc=1)
    ax.grid(False, which='both')
    return fig,ax
```
